```python
import pandas as pd
df = pd.read_csv('../input/dataset/Dataset.csv')
# df = df.drop(['slno'], axis= 1)
concept = df.values[:, :-1]
target = df.values[:,-1]
df.head()
```

```python
def learn(concept, target):
    specific = concept[0].copy()
    general = [["?" for i in range(len(specific))] for i in range(len(specific))]
    
    for i,h in enumerate(concept):
        if target[i] == "yes":
            for x in range(len(specific)):
                if h[x] != specific[x]:
                    general[x][x] = "?"
                    specific[x] = "?"
        if target[i] == "no":
            for x in range(len(specific)):
                if h[x] != specific[x]:
                    general[x][x] = specific[x]
                else:
                    general[x][x] = "?"
                
    indices = [i for i, val in enumerate(general) if val == ['?','?','?','?','?','?']]
    
    for i in indices:
        general.remove(['?','?','?','?','?','?'])
    
    return specific, general
    
```

```python
s_final, g_final = learn(concept, target)
print(f"Final S: {s_final}")
print(f"Final G: {g_final}")
```
